```python
"""
Created: Wed Apr 26 2020
Author: Kumaran B
"""

# A Simple Convolutional Neural Network

# Part 1 - Building the CNN

# Importing the Keras libraries and packages

from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
```

```python
# Initialising the CNN
classifier = Sequential()

# Step 1 - Convolution
classifier.add(Conv2D(16, (3, 3), input_shape = (64, 64, 3), activation = 'relu'))

# Step 2 - Pooling
classifier.add(MaxPooling2D(pool_size = (2, 2)))

# Adding a second convolutional layer
classifier.add(Conv2D(128, (3, 3), activation = 'relu'))

# Adding a second Maxpool layer
classifier.add(MaxPooling2D(pool_size = (2, 2)))

#Flattening the array so that it can be sent to feed forward network
classifier.add(Flatten())

# Step 4 - Full connection
#First Hidden layer
classifier.add(Dense(units = 128, activation = 'relu'))

#Second Hidden layer
classifier.add(Dense(units = 64, activation = 'relu'))

#Output layer
classifier.add(Dense(units = 4, activation = 'softmax'))

# Compiling the CNN
classifier.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
classifier.summary()
```

```python
#Preparing Training and Test images

from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)

test_datagen = ImageDataGenerator(rescale = 1./255)

training_set = train_datagen.flow_from_directory(r'H:\Downloads\INeuron Deep Learning\Data sets\myimagedataset\classes\train_images',
                                                 target_size = (64, 64),
                                                 batch_size = 16)

test_set = test_datagen.flow_from_directory(r'H:\Downloads\INeuron Deep Learning\Data sets\myimagedataset\classes\test_images',
                                            target_size = (64, 64),
                                            batch_size = 16)
```

```python
#Passing train and test images to pipeline

classifier.fit_generator(training_set,
                         epochs = 1000,
                         validation_data = test_set)
```

```python
#Save the model
try:
    classifier.save(r'H:\Downloads\INeuron Deep Learning\Projects\Multi-Classification\model.h5')
except Exception as e:
    print("Error in saving model.", e)
else:
    print("Model saved successfully.")
```

```python
#Prediction using test images

import numpy as np
from keras.preprocessing import image
from pathlib import Path

test_folder = Path(r'H:\Downloads\INeuron Deep Learning\Data sets\myimagedataset\model_test')
class_list = list(training_set.class_indices)

for test_img in test_folder.iterdir():
    test_image = image.load_img(test_img, target_size = (64, 64))
    test_image = image.img_to_array(test_image)
    test_image = np.expand_dims(test_image, axis = 0)
    result = classifier.predict(test_image)
    output = result[0].argmax()
    print(f'The Given file {test_img.name} is identified as {class_list[output]}')
```
